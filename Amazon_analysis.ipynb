**Project Objective :**
The goal of this project is to analyze Amazon product data using Python in order to extract insights related to pricing, ratings, and product categories.

Dataset Description

```python
# This dataset contains 1,465 observations and 16 features,
# where each row represents a single user review for a product listed on Amazon.
# It includes product metadata, pricing information (actual price, discounted price, discount percentage),
# numerical ratings, and textual review content.
```

```python
# Importing the important MODULES
import numpy as np
import pandas as pd
import plotly.express as px 
```

```python
# Reading the dataset
df = pd.read_csv('amazon.csv')
```

2 Clean Data

```python
columns = ['discounted_price', 'actual_price','discount_percentage','rating_count']
chars_to_remove = ['₹', ',', '%']

for col in columns:
    df[col] = df[col].astype(str)

    # remove unwanted characters
    for char in chars_to_remove:
        df[col] = df[col].str.replace(char, ' ', regex=False)

    # convert to float
    df[col] = pd.to_numeric(df[col] , errors = 'coerce')

    # fill NaN with 0 (or mean/median – your choice)
    df[col] = df[col].fillna(0)

    # finally convert to integer
    df[col] = df[col].astype(int)
```

```python
df.info()
```

3 Visualization

```python
# -----The chart shows how ratings are distributed in the dataset-----

# Converting rating values ​​to numbers
df['rating'] = pd.to_numeric(df['rating'], errors = 'coerce')

# Fill in the missing values ​​with an median
df['rating'] = df['rating'].fillna(df['rating'].median())

# Box Plot
fig = px.box(
    df,
    x = 'rating',
    title = "Rating Distribution",
)

fig.update_layout(
    xaxis_title = "Rating (2-5)",
    showlegend = False
)

fig.show()
```

```python
#-----Rating Count Distribution-----

rating_counts = df['rating'].value_counts().sort_index()

fig = px.bar(
    x = rating_counts.index,
    y = rating_counts.values,
    title = "Rating Count Distribution",
    labels = {
        'x': 'Rating',
        'y': 'Count of Ratings'
    },
)

fig.update_layout(
    xaxis_title = "Rating (1-5)",
    yaxis_title = "Count of Ratings"
)

fig.show()
```

```python
#-----Top 5 categories in sales-----

# Extracting the Category name from the Category column
df['category'] = df['category'].str.extract(r'^([^|]+)')

# Select the top 5 categories
top5_categories = df['category'].value_counts().head(5)

fig = px.bar(
    x = top5_categories.index,
    y = top5_categories.values,
    color = top5_categories.index,
    text = top5_categories.values
)

fig.update_layout(
    xaxis_title = "Top 5 Categories",
    yaxis_title = "Number of Sales for Each Product"
)

fig.update_traces(
    textposition = 'outside',
    hovertemplate = 
    'Category : %{x}<br>'
    'Sales : %{y}<extra></extra>' 
)

fig.show()
```

```python
#-----Shows the relationship between price and rating-----

fig = px.scatter(
    df,
    x = 'discounted_price',
    y = 'rating',
    title = 'Discounted Price vs Rating',
)

fig.update_layout(
    xaxis_title = "Discounted Price ($)",
    yaxis_title = "Rating"
)

fig.show()
```

```python
#-----Compares discount percentages across different product categories-----

fig = px.box(
    df,
    x = 'category',
    y = 'discount_percentage',
    title = 'Discount Percentage by Category',
)

fig.update_layout(
    xaxis_title = "Category names",
    yaxis_title = "Discount (%)"
)

fig.show()
```

```python
#-----Shows the correlation between price, discount, rating, and rating count-----

columns = [
    'discounted_price',
    'actual_price',
    'discount_percentage',
    'rating',
    'rating_count'
]

corr_matrix = df[columns].corr()

fig = px.imshow(
    corr_matrix,
    text_auto = '.2f',
    title = 'Correlation Matrix of Pricing and Rating Variables',
    color_continuous_scale = 'Blues',
    width = 800,
    height = 600
)

fig.update_traces(
    hovertemplate=
    'X: %{x}<br>'
    'Y: %{y}<br>'
    'Correlation: %{z:.2f}<extra></extra>'
)

fig.show()
```
